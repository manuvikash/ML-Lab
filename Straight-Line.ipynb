```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./datasets/linear-reg.csv')
print(data.shape)
data.head()
data = np.array(data)
np.random.shuffle(data)
```

```python
trainingSplit = 0.8
trainingSize = int(trainingSplit * len(data))
XTrain = data[:trainingSize,0]
YTrain = data[:trainingSize,1]
XTest = data[trainingSize:,0]
YTest = data[trainingSize:,1]
```

image.png

```python
meanX = np.mean(XTrain)
meanY = np.mean(YTrain)

slope = np.sum((XTrain - meanX) * (YTrain - meanY))/np.sum((XTrain - meanX) ** 2)
intercept = meanY - (slope * meanX)
print(slope)
print(intercept)
```

```python
#plot the line
max_x = np.max(XTrain) + 100
min_x = np.min(XTrain) - 100
x = np.linspace(min_x, max_x, 1000)
y = slope * x + intercept

plt.plot(x, y, color = 'blue', label = 'regression line')
plt.scatter(XTrain, YTrain, color = 'red', label = 'scatter plot')
plt.legend()
plt.show()
```

```python
#Accuracy with training data
pred = slope * XTrain + intercept
error = np.sum((YTrain - pred) ** 2)
print(error)
```

```python
#plot with test data
max_x = np.max(XTest) + 100
min_x = np.min(XTest) - 100
x = np.linspace(min_x, max_x, 1000)
y = slope * x + intercept

plt.plot(x, y, color = 'blue', label = 'regression line')
plt.scatter(XTest, YTest, color = 'red', label = 'scatter plot')
plt.legend()
plt.show()
```

```python
#Accuracy with test data
pred = slope * XTest + intercept
error = np.sum((YTest - pred) ** 2)
print(error)
```
